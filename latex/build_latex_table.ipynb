```python
# preprocess
datasets = [d.replace('_','').replace('-','') for d in datasets]
df_all=df_all*100
df_all.insert(0,'datasets',datasets)
max_scores = []
import numpy as np
for i in range(df_all.shape[1]):
    if i!=0:
        df_all.iloc[:,i]=np.round(df_all.iloc[:,i].values,2)

for i in range(df_all.shape[0]):
    scores = df_all.iloc[i,1:].values
    scores=scores[np.arange(0,df_all.shape[1]-1,2)]
    max_scores.append(max(scores))

df_all = pd.DataFrame(df_all,dtype=str)
for i in range(df_all.shape[0]):
    for j in range(1, df_all.shape[1], 2):
        if float(df_all.iloc[i,j])==max_scores[i]:
            df_all.iloc[i,j]='\\bf{'+df_all.iloc[i,j]
            df_all.iloc[i,j+1]=df_all.iloc[i,j+1]+'}'
        
for i in range(df_all.shape[1]):
    if i==0:
        df_all.iloc[:,i]=df_all.iloc[:,i]+' &'
    else:
        if i%2==1:
            df_all.iloc[:,i]=' $ '+df_all.iloc[:,i]+' \pm '
        else:
            if i==df_all.shape[1]-1:
                df_all.iloc[:,i]+=' $\\\\'
            else:
                df_all.iloc[:,i]+=' $ &'
    
df_all.head()
```

```python
df_all.to_csv('auc_latex.csv',index=None)
```
